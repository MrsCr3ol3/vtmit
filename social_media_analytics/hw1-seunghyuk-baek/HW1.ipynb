```python
%matplotlib inline
import networkx as nx;
import matplotlib.pyplot as plt;
import warnings

warnings.filterwarnings("ignore")

G = nx.Graph()

G = nx.read_edgelist("facebook_combined.txt")

for e in G.edges():
    print e
```

```python
#average degree of edges
a=0
b=0

for n in G.nodes():
    a=a+G.degree(n)
    b=b+1
print a, b, a/b
```

```python
#number of connectd component
print nx.number_connected_components(G)
```

```python
giant = max(nx.connected_component_subgraphs(G), key=len)
print nx.diameter(giant)
```

```python
import operator 

degree_cent = nx.degree_centrality(G)
sorted_dc = sorted(degree_cent.items(), key=operator.itemgetter(1),reverse=True)

print sorted_dc
```

```python
nx.density(G)
```

```python
import numpy as np
import matplotlib.mlab as mlab

# the histogram of the data
plt.hist(sorted_dc, bins = 100)
plt.title('degree centrality plot')
plt.ylabel('frequency')
plt.xlabel('degree')
```

```python
#draw graph
nx.draw(G)
```

```python
degree_sequence=sorted(nx.degree(G).values(),reverse=True) # degree sequence
#print "Degree sequence", degree_sequence
dmax=max(degree_sequence)

plt.loglog(degree_sequence,'b-',marker='o')
plt.title("Degree rank plot")
plt.ylabel("degree")
plt.xlabel("rank")
```

```python
#centralization

sumof = 0;
length = len(degree_sequence)
for i in degree_sequence:
    sumof = sumof + (dmax - i)

c = (sumof * 100) / ((length-1)*(length-2))
c, sumof, len(degree_sequence), dmax
```
