```python
%matplotlib inline
import sklearn
#import graphviz
#import sklearn.tree
import numpy as np
#from StringIO import StringIO
import matplotlib.pyplot as plt
from sklearn import neighbors
from sklearn import datasets, linear_model
```

```python
import re

file1 = open('file1.csv','r')

file1list = list()

for line in file1:
    newLine = line.rstrip()
    newLine = re.sub(r'(\".*)\,(.*\")',r'\1 \2',newLine)
    x = np.array(newLine.split(","))
    file1list.append(x)

import pandas as pd

df1 = pd.DataFrame(file1list[1:], columns = file1list[0])

print (df1)

file1.close()
```
