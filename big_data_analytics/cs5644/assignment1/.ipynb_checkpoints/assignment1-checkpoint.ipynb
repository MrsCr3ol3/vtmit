1) Define a function is_palindrome() in Python that recognizes palindromes (i.e. words that look the same written backwards). For example, is_palindrome("radar") should return True. is_palindrome("CS5644") should return False. 

```python
def is_palindrome(param):
rev_param = param[::-1]   
    if param == rev_param:
        return True
    else:
        return False
```

2)	(15 points) In cryptography, a Caesar cipher is a very simple encryption technique in which each letter in the plain text is replaced by a letter some fixed number of positions down the alphabet. For example, with a shift of 3, A would be replaced by D, B would become E, and so on. The method is named after Julius Caesar, who used it to communicate with his generals. ROT-13 ("rotate by 13 places") is a widely used example of a Caesar cipher where the shift is 13. In Python, the key for ROT-13 may be represented by means of the following dictionary:

```python
from string import maketrans

def rot13Encoder(param):
    lowCaseOri = 'abcdefghijklmnopqrstuvwxyz'
    lowCaseEncypt = 'nopqrstuvwxyzabcdefghijklm'
    upCaseOri = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    upCaseEncrypt = 'NOPQRSTUVWXYZABCDEFGHIJKLM'
    
    encryptLow = maketrans(lowCaseOri, lowCaseEncypt)
    encryptUp = maketrans(upCaseOri, upCaseEncrypt)
    
    print param.translate(encryptLow).translate(encryptUp)
```

(Data exploration)

The following two questions involve the breast cancer dataset from the UCI machine learning repository. This dataset contains 700 instances, each with 8 features. The class information indicates whether the instance is benign or malignant. You will find all the necessary data in data.csv. A jupyter notebook is also provided along with the assignment that has a helper function to read data from the file into a NumPy array. 


```python
#helper function to load data into files.
def _loadData():
    import numpy as np
    ids=list()   #This will indicate the unique id of the sample.
    clump_thickness=list()  #This property is a number between 1 - 10
    single_epithelial_cell_size=list() #long variable names are usually bad practice but this has been done to make things clear.
    target_class=list()   #This depicts whether the tumor is benign (class = 2) or malignant (class = 4). 
    seen=set()
    
    with open("data.csv") as f:
        for idx,line in enumerate(f):
            if idx==0:
                continue   #we ignore the first line as it's just the header, so we just `continue` which goes to the next line in the file.
            
            else:
                _id,_ct,_secs,_cl=line.strip().split(",")  #we are using the strip() function to remove any leading and trailing whitespace.
                seen.add(_id)
                ids.append(_id)   
                clump_thickness.append(int(_ct)) #notice the int() in the append function, this is done to "cast" the string values which is how the numbers are read from the file, to integers
                single_epithelial_cell_size.append(int(_secs))
                target_class.append(int(_cl))
    
    return np.array(ids),np.array(clump_thickness),np.array(single_epithelial_cell_size),np.array(target_class)

"""
1. The _loadData() function returns 4 numpy arrays (unlike java, python can return multiple values, 
here we are returning 4 arrays).

2. The unique sample ids are stored in the `ids` array, 
3. The clump thickness is stored in the `clump_thickness` array
4. The single epithelial cell size feature is stored in the `single_ep_cell_size` array and 
5. The class label for each sample whether `malignant` or `benign` is stored in the `benign_or_malignant` array.

"""

ids,clump_thickness,single_ep_cell_size,benign_or_malignant=_loadData()   #this is how you invoke the helper function.
```

```python
#import libararies here.
%matplotlib inline    
import numpy as np
import matplotlib.pyplot as plt   
from sklearn import datasets 
import seaborn as sns   
sns.set()
```

3) Create a stacked histogram showing the clump thickness for malignant and benign tumors. With the malignant class colored red and the benign class colored blue.


```python
#your code here 
fig,ax=plt.subplots(figsize=(12,8))
tmp=list()
colors = ['blue', 'red']

for i in range(2):
    indices=np.where(benign_or_malignant==((i+1)*2))
    tmp.append(clump_thickness[indices])

ax.hist(tmp,bins=10,stacked=True, color = colors, edgecolor='black', linewidth=1.2)   

plt.title("Clup Thickness Histogram",fontsize=30)
plt.xlabel("Size",fontsize=24)
plt.ylabel("Number of Occurrences",fontsize=24)
ax.legend(['Benign','Malignant'],loc='upper right',fontsize=24)
```

4) Create a scatter plot to visualize the relationship between clump thickness and single epithelial cell size and color each instance red or blue based on whether it's malignant or benign respectively.


```python
#your code here
fig,ax=plt.subplots(figsize=(12,8))
colors=['b','r']  #create a simple colors list with 3 colors blue,green,red indicated by 'b','g','r' respectively. 
for i in range(2): #since there are 3 types of flowers in our dataset
    indices=np.where(benign_or_malignant==((i+1)*2))
    ax.scatter(clump_thickness[indices],single_ep_cell_size[indices],s=(i+1)*25,c=colors[i], alpha = 0.2)   #basic scatter plot with sepal_length as x-axis and sepal_width as y-axis.

#title,axis labels and legend
plt.title("Clump Thickness vs. Single ep Cell Size ",fontsize=30)
plt.xlabel("Clump Size",fontsize=24)
plt.ylabel("EP Cell Size",fontsize=24)
ax.legend(['Benign','Malignant'],loc='upper right',fontsize=24,bbox_to_anchor=(0.43, 0.92, 1., .102),mode="tight")
```
